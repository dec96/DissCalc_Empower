```python
#Program returns Empower data in a format that can be copied to DissCalc

import string
from openpyxl import Workbook
from openpyxl import load_workbook


import warnings
warnings.filterwarnings("ignore")

def divide_list(l, n, x):
    for i in range(x, len(l), n): yield l[i:i+n]

def standardsValue(standard, listStd):
    print("-"*26)
    for X in range(1, int(len(listStd)+1)):
        print(standard + " " + str(X) + ": " + listStd[int(X-1)])
        
    suM = 0
    for x in listStd:
        if x != "": suM += float(x)
    if len(listStd) != 0:
        average = suM/ int(len(listStd))
        if standard.lower() != "blank":
            print("\n" + standard + " Average : " + str(average))

def valuesArranged(values, n):

    dissValues = list(divide_list(values, n, 0))
    print("")
    for i in dissValues:
        x = " ".join(i)
        print(x)


def identifyLabel(lines, position):
    listBlank, listCal, listRec, listSample = [],[],[],[]

    #print(lines)

    listCal = ([(i[position]) for i in lines if "Cal" in str(i[3])])
    listBlank = ([i[position] for i in lines if "blank" in (str(i[3]).lower())])
    listRec = ([i[position] for i in lines if "rec" in (str(i[3]).lower())])

    
    for i in lines:
        if not i[position] in listBlank and not i[position] in listCal and not i[position] in listRec:
            listSample.append(i[position])
    #print(listCal, listBlank, listRecovery, listSample)

    return (listBlank, listCal, listRec, listSample)

def importToDissCalc(user, listCal, listRec, listSamples, n, values, massCal, massRec):
    print("\n\nLoading DissCalc 2 Template.xlsm and updating fields... please wait...")
    wb = load_workbook('C:\\Users\\' + user + '\Desktop\\DissCalc 2 Template.xlsm', keep_vba = True)
    
    sheet = wb.get_sheet_by_name("DissCalc report")
    #print(listRec)
    calAvg = sum(map(int,listCal))/len(listCal)
    recAvg = sum(map(int,listRec))/len(listRec)
    sheet["B53"] = calAvg
    sheet["B57"] = recAvg
    if massCal != " ":
        sheet["B51"] = float(massCal)
        sheet["B55"] = float(massRec)
    column = list(string.ascii_uppercase[4:int(4+n)])
    dissValues = list(divide_list(values, n, 0))
    ################################################
    for i in dissValues:
        for x in i:
                sheetCoords = str(str(column[i.index(x)]) + str(29 + dissValues.index(i))) 
                sheet[sheetCoords] = float(x)
    ################################################
    sheet = wb.get_sheet_by_name("Standards report")
    if massCal != " ":
        sheet["C23"] = float(massCal)
        sheet["H23"] = float(massRec)
    for i in listCal:
        sheetCoords = str("C" + str(38 + listCal.index(i)))
        sheet[sheetCoords] = float(i)
                  
    for i in listRec:
        sheetCoords = str("H" + str(38 + listRec.index(i)))
        sheet[sheetCoords] = float(i)
    ######################################################
    newFileName = str("C:\\Users\\" + user + "\Desktop\\" + str(input("\nNew file name: ")) + ".xlsm")
    print("\n\nSaving updated copy of DissCalc 2 Template as " + str(newFileName) + " on your desktop...")
    wb.save(newFileName)
    print("\nDissCalc 2 Template has been updated and saved as " + str(newFileName) + " on your desktop.")

    

def disscalcEmpower():
    print("\nExport your Empower Data using the Analytical Export Method")
    print("\nBefore continuing, make sure that file is in format 'Analytical_Export(code)'")

    #fileName = str("\\\semldsw2fil275.rd.astrazeneca.net\Laban\UK_Empower_Export\Analytical Export\Analytical_Export" + str(input("\nExport code: "))+ ".txt")

    user = str(input("\nEnter your username (PRID): "))
    fileName = str("C:\\Users\\" + user + "\Desktop\Analytical_Export" + str(input("\nExport code: "))+ ".txt")
    #"C:\Users\knxk542\Desktop\Analytical_Export27436.txt"

    n = int(input("\nThe value of n: "))
    
    if str(input("\nDo you wish to enter standard weights? (Y/N) ")) == "Y":
            massCal = str(input("\nEnter the mass of calibration standard (mg): "))
            massRec = str(input("\nEnter the mass of recovery standard (mg): "))
    else:
            massCal, massRec = " ", " "

    with open(fileName, "r") as file:
        data = ("\t".join(file.readlines()))
        dataSplit = (data.split("\t"))
        dataRows = list(divide_list(dataSplit, 13, 0))
        #print(dataRows)

        listBlank, listCal, listRec, listSample = identifyLabel(dataRows[1:], 9)
        #print(listBlank, listCal, listRec, listSample)

        importToDissCalc(user, listCal, listRec, listSample, n, listSample, massCal, massRec)
        #standardsValue("Blank", listBlank), standardsValue("Calibration", listCal), standardsValue("Recovery", listRec)
        #valuesArranged(listSample, n)

replayCode = True        
while replayCode:
    print("\nBefore continuing, make sure Excel is up to date and no updates are needed.")
    disscalcEmpower()
    print("\nWhen opening file, click 'yes' to repairs.")
    if str(input("\n\nDo you wish to replay the programme? (Y/N) ")).lower() == "n": replayCode=False
    print("-" * 50)
    
    

```
